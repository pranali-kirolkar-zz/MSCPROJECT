```python
#importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
#Sales of Summer clothes Dataset
Dataset = pd.read_csv('//content/Sales Summer Clothes....csv') 
Dataset
```

```python
pd.set_option('display.max_columns',None)
```

```python
duplicate=Dataset[Dataset.duplicated()]
duplicate
```

```python
Dataset=Dataset.fillna(method='ffill')
Dataset
```

```python
Dataset.columns
```

```python
Dataset['compare prices '].value_counts().plot(kind='pie',autopct='%1.2f%%',figsize=(10,8))
plt.title("Do you compare prices to make purchase decisions?")
plt.show()
```

```python
Dataset['any worst experience'].value_counts().plot(kind='pie',autopct='%1.2f%%',figsize=(10,8))
plt.title("Is there any worst experience in clothes sale?")
plt.show()
```

```python
Dataset[' prefer purchasing '].value_counts().plot(kind='pie',autopct='%1.2f%%',figsize=(10,8))
plt.title("Where do you prefer purchasing clothes?")
plt.show()
```

```python
Dataset['time_prefer '].value_counts().plot(kind='pie',autopct='%1.2f%%',figsize=(10,8))
plt.title("Which time do you prefer purchasing Clothes?")
plt.show()
```

```python
Dataset['Rating '].value_counts().plot(kind='pie',autopct='%1.2f%%',figsize=(10,8))
plt.title("Rating")
plt.show()
```

```python
Dataset['Price of clothes'].value_counts().plot(kind='pie',autopct='%1.2f%%',figsize=(10,8))
plt.title("Price of the Clothes :")
plt.show()
```

```python
Dataset['buy clothes'].value_counts().plot(kind='pie',autopct='%1.2f%%',figsize=(10,8))
plt.title("How often do you buy clothes in the sale?")
plt.show()
```

```python
Dataset['Brand of clothes'].value_counts().plot(kind='bar',figsize=(15,8))
plt.title("Which Brand of clothing have you purchased?")
plt.show()
```

```python
Dataset['Type of clothes '].value_counts().plot(kind='bar',figsize=(15,8))
plt.title("Which type of clothing have you purchased recently?")
plt.show()
```

```python
Dataset['Age'].value_counts().plot(kind='bar',figsize=(15,8))
plt.title("How old are you?")
plt.show()
```

```python
Dataset['Price of clothes'].value_counts().plot(kind='bar',figsize=(10,8))
plt.title("Price of the Clothes :")
plt.show()
```
